### Activation Functions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Binary Step

```python
def binaryStep(x):
    return np.heaviside(x,1)

x=np.linspace(-10,10)
plt.plot(x,binaryStep(x))
plt.axis('tight')
plt.title('Activation function:Binary Step')
plt.grid(True)
plt.show()
```

### Sigmoid

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x=np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('Activation function : Sigmoid')
plt.grid(True)
plt.show()
```

### Tanh (Hyperbolic)

```python
def tanh_fun(x):
    return np.tanh(x)

x=np.linspace(-10,10)
plt.plot(x,tanh_fun(x))
plt.axis('tight')
plt.title('Activation function : tanh(hyperbolic)')
plt.grid(True)
plt.show()
```

### Softmax

```python
def Softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)

x=np.linspace(-5,5)
plt.plot(x,Softmax(x))
plt.axis('tight')
plt.title('Activation function : Softmax')
plt.grid(True)
plt.show()
```

### Exponential Linear Unit

```python
def expnlu(x):
    return 1.0*(np.exp(x)-1)

x=np.linspace(-10,10)
plt.plot(x,expnlu(x))
plt.axis('tight')
plt.title('Activation function :Exponential Linear Unit')
plt.grid(True)
plt.show()
```

### Relu

```python
def relu(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1

x=np.linspace(-10,10)
plt.plot(x,relu(x))
plt.axis('tight')
plt.title('Activation function :Relu')
plt.grid(True)
plt.show()
```

### Leaky Rule Function

```python
def lrf(x):
    x1=[]
    for i in x:
        if i>=0:
            x1.append(i)
        else:
            x1.append(0.01*i)
    return x1

x=np.linspace(-5,5)
plt.plot(x,lrf(x))
plt.axis('tight')
plt.title('Activation Function :Leaky Rule Function ')
plt.grid(True)
plt.show()
```

### Rectified Linear Unit (ReLU)

```python
def rectf(x):
    x1 = []
    for i in x:
        if i >= 0:
            x1.append(i)
        else:
            x1.append(0)
    return x1

x = np.linspace(-10, 10)
plt.plot(x, rectf(x))
plt.axis('tight')
plt.title('Activation function: Rectified Linear Unit (ReLU)')
plt.grid(True)
plt.show()
```
